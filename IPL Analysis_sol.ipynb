### ----------------------------TASK 3 ----------------------------------#
##### Display the Top 10 players interms of high strike rate 
###### (Note that a player must atleast scored 1000 runs to be in this list)

```python
df_filter=df_runs[df_runs['Batsman_Scored'] >=1000]
df_filter.head()
```

```python
ser = ball_by_ball['Striker_Id'].value_counts()
df2 = pd.DataFrame(ser)
df2.reset_index(inplace = True)
df2.rename(columns = {'Striker_Id':'No_of_BallsFaced'},inplace = True)
df2.rename(columns = {'index':'Striker_Id'},inplace = True)
df2.head()
```

```python
df_SR = pd.merge(df_filter,df2,on = 'Striker_Id')
df_SR['Strike_rate'] = (df_SR['Batsman_Scored'] / df_SR['No_of_BallsFaced']) * 100
df3 = df_SR.merge(player,left_on = 'Striker_Id',right_on = 'Player_Id')
df3[['Striker_Id','Player_Name','Batsman_Scored','Strike_rate']].sort_values(by = 'Strike_rate',ascending = False).head(10)
```

### ----------------------------TASK 5 ----------------------------------#
  
##### Display the Top 10 players interms of Economic Rate 

```python
# Solution

ball_by_ball=pd.read_excel('Ball_by_Ball.xlsx')
player = pd.read_excel('Player.xlsx')

ball_by_ball.head()
```

```python
player.head()
```

```python
balls_count = ball_by_ball.groupby('Bowler_Id')[['Batsman_Scored']].count().reset_index().sort_values(by = 'Batsman_Scored',ascending = False)
balls_count.head()
```

```python
runs_conceded  = ball_by_ball.groupby('Bowler_Id')[['Batsman_Scored']].sum().reset_index().sort_values(by = 'Batsman_Scored',ascending = False)
runs_conceded.head()
```

```python
bowler_details = pd.merge(balls_count,runs_conceded,on = 'Bowler_Id')
bowler_details = bowler_details.rename(columns = {'Batsman_Scored_x':'Deliveries_Bowled','Batsman_Scored_y':'Runs_conceded'})
bowler_details.head()
```

```python
bowler_details['Overs_bowled'] = round(bowler_details['Deliveries_Bowled'] / 6,1)
bowler_details.drop('Deliveries_Bowled',axis = 1,inplace = True)
bowler_details['Economy_rate'] = round(bowler_details['Runs_conceded'] / bowler_details['Overs_bowled'],2)
bowler_details.sort_values('Economy_rate',ascending = True).head(10)
```

```python
# Since economy rate as such does not make much sense, we shall put a criteria that the bowler must have bowled atleast 30 
# overs. So that economy rate can be meaning
```

```python
bowler_details_refined = bowler_details[bowler_details['Overs_bowled'] >30]
final_bowler_details = pd.merge(bowler_details_refined,player,left_on = 'Bowler_Id',right_on = 'Player_Id')
```

```python
final_bowler_details[['Bowler_Id','Player_Name','Bowling_Skill','Overs_bowled','Economy_rate']].sort_values('Economy_rate',ascending = True).head(10)
```

### ----------------------------TASK 6 ----------------------------------#
#### Is that the CSK is a better team when they bat second ?

##### Determing W/L ratio batting first

```python
match=pd.read_excel('Match.xlsx')
match_csk = match[(match['Team_Name_Id'] == 3) | (match['Opponent_Team_Id'] == 3)] # For choosing matches played by CSK
cond_1 = (match_csk['Toss_Winner_Id'] == 3) & (match_csk['Toss_Decision'] == 'bat')
cond_2 = (match_csk['Toss_Winner_Id'] != 3) & (match_csk['Toss_Decision'] == 'field')
match_csk_bat_first = match_csk[(cond_1) | (cond_2)] # Total no of games which has been playes batting first = 77
match_csk_bat_first[match_csk_bat_first['Match_Winner_Id'] == 3]['Match_Winner_Id'].value_counts() # No.of games won batting first = 45
```

```python
batting_first_lost =len(match_csk_bat_first) - match_csk_bat_first[match_csk_bat_first['Match_Winner_Id'] == 3]['Match_Winner_Id'].value_counts()
```

```python
batting_first_won = match_csk_bat_first[match_csk_bat_first['Match_Winner_Id'] == 3]['Match_Winner_Id'].value_counts()
W_L_BattingFirst = batting_first_won / batting_first_lost 
W_L_BattingFirst
# W/L_BattingFirst = 1.40625
```

##### Determining W/L ratio batting second

```python
cond_1 = (match_csk['Toss_Winner_Id'] == 3) & (match_csk['Toss_Decision'] == 'field')
cond_2 = (match_csk['Toss_Winner_Id'] != 3) & (match_csk['Toss_Decision'] == 'bat')
match_csk_bat_second = match_csk[(cond_1) | (cond_2)]  # Total no of games which has been playes batting second = 54
match_csk_bat_second[match_csk_bat_second['Match_Winner_Id'] == 3]['Match_Winner_Id'].value_counts()
```

```python
batting_second_lost =len(match_csk_bat_second) - match_csk_bat_first[match_csk_bat_first['Match_Winner_Id'] == 3]['Match_Winner_Id'].value_counts()
batting_second_lost
```

```python
batting_second_won = match_csk_bat_second[match_csk_bat_second['Match_Winner_Id'] == 3]['Match_Winner_Id'].value_counts()
W_L_BattingSecond = batting_second_won / batting_second_lost 
W_L_BattingSecond
# # W/L_BattingSecond = 1.7
```

```python
# W/L Batting first = 1.40625
# W/L Batting second = 1.7

# We can clearly say that batting second has been doing good for team CSK
```
